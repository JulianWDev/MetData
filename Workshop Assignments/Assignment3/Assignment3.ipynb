```python
obj_file = open("./10-498-598-LoD12-3D.obj", "r")
print(obj_file)
```

```python
data = []
for i in obj_file:
    if i[0] == 'v':
        g = i.split(' ')
        del g[0]
        for j in range(len(g)):
            g[j] = float(g[j])
        data.append(g)
        
print(data)
```

```python
coordinate_x = []
coordinate_y = []
coordinate_z = []

for i in range(len(data)):
    coordinate_x.append(data[i][0])
    coordinate_y.append(data[i][1])
    coordinate_z.append(data[i][2])

xdiff = max(coordinate_x) - min(coordinate_x)
ydiff = max(coordinate_y) - min(coordinate_y)
zdiff = max(coordinate_z) - min(coordinate_z)

volume = xdiff * ydiff * zdiff

print(volume)
```
