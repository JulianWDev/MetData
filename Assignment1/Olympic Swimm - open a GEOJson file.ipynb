```python
import pandas as pd
import requests
import json

import geopandas as gpd
import osmnx as ox
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon

%matplotlib inline
```

```python
# call the url link which stores the geo data file
url_swimm_loc = "https://maps.amsterdam.nl/open_geodata/geojson_latlng.php?KAARTLAAG=ZWEMWATER&THEMA=zwemwater"
```

```python
# download the data in the url
recall_link_swimm_loc = requests.get(url_swimm_loc)
```

```python
# load the geographic data into a data frame
geo_dataframe_swimm_loc = gpd.read_file(recall_link_swimm_loc.text)
```

```python
# view the first few line of the code to make sure that the data has the right stucture
print(geo_dataframe_swimm_loc.head())
```

```python
# plot the geogrpahic data
fig, ax = plt.subplots(figsize=(200,200))
# standardize
ax.set_aspect('equal')
#set the main colour
ax.set_facecolor('#ED1C24')

#plot my data
geo_dataframe_swimm_loc.plot(ax=ax, marker = 'o', markersize = 1500, color='white')

#save file as an image to my files
plt.savefig("zwimplaces amsterdam.png")

#show my data
plt.show()
```
