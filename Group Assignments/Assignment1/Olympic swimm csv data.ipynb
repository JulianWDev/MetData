```python
#reads the data to a dataframe and seperates by whatever value is defined by setting sep otherwise its automatically a comma
df = pd.read_csv('ZWEMWATER 1.csv', sep = ";")
```

```python
# This code is simpler but works as well. By using pandas quite a bit of the possible editing is avoided which makes it easier to use the data later
csv_data = []
my_file = open("ZWEMWATER 1.csv", 'r')
for line in my_file:
    data = line.split(";")
    csv_data.append(data)
my_file.close()
print(max(csv_data), min(csv_data))
csv_data
```
